# Effect of Changing the Number of Training Points
While keeping the other variables the same, we can change the number of training data points $N$ to see how it affects the KNN regression. We repeat the same process with smaller and larger $N$ and observe the effects of changing them.

# Smaller Number of Training Points $N$
$N=500$

```python
from matplotlib import pyplot as plt
import numpy as np

N = 500
sigma = 1
a = 2 

x_train = a*np.random.rand(N)
eps = np.random.normal(0, sigma, N)
y_train = 2*x_train + np.sin(2*np.pi*x_train) + eps;
```

```python
x = np.linspace(0, a, 100) 
y = 2*x + np.sin(2*np.pi*x) 

figure, axis = plt.subplots()
axis.set_title('Training Data: N = 500')
plt.scatter(x_train,y_train, c='g')
plt.plot(x, y, c='b')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

K = 25

model = KNeighborsRegressor(n_neighbors = K)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function: N = 500'.format(K))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```

# Observation
Smaller number of training points results in smaller oscillation of the KNN regression graph. However, with a small training data set, the prediction becomes less accurate. For instance, the horizontal lines (slope = $0$) at the end points of the graphs become clearer as $N$ decreases.

With larger $N$, KNN regression graphs become more similar to the theoretical regression graph, and their oscillation varies more.
