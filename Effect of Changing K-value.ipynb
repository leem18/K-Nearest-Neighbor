# Effect of Changing K-value in KNN
While keeping the other variables the same, we can change $K$-value to see how it affects the KNN regression. We repeat the same process with larger and smaller $K$-values to observe the effect of changing $K$.

```python
from matplotlib import pyplot as plt
import numpy as np

N = 1000
sigma = 1 
a = 2 

x_train = a*np.random.rand(N)
eps = np.random.normal(0, sigma, N)
y_train = 2*x_train + np.sin(2*np.pi*x_train) + eps;
```

```python
x = np.linspace(0, a, 100) 
y = 2*x + np.sin(2*np.pi*x) 

figure, axis = plt.subplots()
axis.set_title('Training Data')
plt.scatter(x_train,y_train, c='g')
plt.plot(x, y, c='b')
plt.show()
```

# Larger K-Values
$K=30$

```python
from sklearn.neighbors import KNeighborsRegressor

K30 = 30 

model = KNeighborsRegressor(n_neighbors = K30)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function'.format(K30))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```

$K=40$

```python
K40 = 40 

model = KNeighborsRegressor(n_neighbors = K40)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function'.format(K40))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```

# Smaller K-Values
$K=20$

```python
K20 = 20 

model = KNeighborsRegressor(n_neighbors = K20)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function'.format(K20))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```

$K=10$

```python
K10 = 10 

model = KNeighborsRegressor(n_neighbors = K10)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function'.format(K10))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```

# Observation
As $K$ gets smaller, the oscillation of the regression graphs varies more. However, with larger $K$-values, we have the regression graphs which are much more similar with the theoretical regression graph (blue line).

In addition, with the larger $K$-values, we can see that the slope of $f(x)$ near $x=0$ and $x=2$ becomes closer to $0$. We can observe the horizontal lines at the extreme points of the graphs.
