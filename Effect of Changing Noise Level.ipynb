# Effect of Changing Noise Level
While keeping the other variables the same, we can change the noise levels $\sigma$ to see how it affects the KNN regression. We repeat the same process with smaller and larger $\sigma$ and observe the effects of changing them.

# Smaller Noise Level $\sigma$
$\sigma=0.5$

```python
from matplotlib import pyplot as plt
import numpy as np

N = 1000
sigma1 = 0.5
a = 2 

x_train = a*np.random.rand(N)
eps = np.random.normal(0, sigma1, N)
y_train = 2*x_train + np.sin(2*np.pi*x_train) + eps;
```

```python
x = np.linspace(0, a, 100) 
y = 2*x + np.sin(2*np.pi*x) 

figure, axis = plt.subplots()
axis.set_title('Training Data: sigma = 0.5')
plt.scatter(x_train,y_train, c='g')
plt.plot(x, y, c='b')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

K = 25

model = KNeighborsRegressor(n_neighbors = K)
model.fit(x_train.reshape(-1, 1), y_train)


y_predicted = model.predict(x.reshape(-1,1))

figure, axis = plt.subplots()
axis.set_title('{}-Nearest Neighbor Regression Function: sigma = 0.5'.format(K))
plt.plot(x, y, c='b')
axis.plot(x, y_predicted, 'r')
plt.show()
```
